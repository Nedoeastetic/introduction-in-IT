# Глобальные, локальные и нелокальные переменные

В Python переменная, объявленная вне
функции или в глобальной области
видимости, называется глобальной
переменной. К глобальной
переменной можно получить доступ
как внутри, так и вне функции.
На примере, как в Python создается
глобальная переменная.


````python
x= 'глобальная переменная'

def foo():
print('х внутри функции:',x)
foo()
print('х  вне функции:',x)
````

В приведенной выше программе мы
создали глобальную переменную x и
задали функцию foo(), которая выводит
на экран значение x. В коде программы
мы вызвали функцию foo(), которая
напечатала значение x внутри
функции. Как вы видите, оно совпадает
со значением x вне функции.


Переменная, объявленная внутри тела
функции или в локальной области
видимости, называется локальной
переменной.

````python

def foo():
  y= 'локальная переменная'

foo()
print(y)
````

Python выдает ошибку, потому что мы
пытаемся получить доступ к локальной
переменной y в глобальной области
видимости. Так делать нельзя:
локальная переменная y «существует»
только внутри функции foo()

---
Как использовать глобальные и
локальные переменные в одной
программе?


````python
x = "глобальная переменная"
def foo():
global x
y = "локальная переменная"
x = x * 2
print(x)
print(y)
foo()
````

В приведенном выше программе мы
объявили глобальную переменную x и

оператор умножения, чтобы изменить
глобальную переменную x, и вывели на
экран значения переменных x и y.

# Базовые алгоритмы. Сортировка Пузырьком
Этот простой алгоритм выполняет итерации по
списку, сравнивая элементы попарно и меняя их
местами, пока более крупные элементы не
«всплывут» в начало списка, а более мелкие не
останутся на «дне».
Сначала сравниваются первые два элемента
списка. Если первый элемент больше, они
меняются местами. Если они уже в нужном
порядке, оставляем их как есть. Затем переходим к
следующей паре элементов, сравниваем их
значения и меняем местами при необходимости.
Этот процесс продолжается до последней пары
элементов в списке.
При достижении конца списка процесс
повторяется заново для каждого элемента. Это
крайне неэффективно, если в массиве нужно
сделать, например, только один обмен. Алгоритм
повторяется n² раз, даже если список уже
отсортирован.
Чтобы остановить алгоритм по окончании
сортировки, нужно ввести переменную-флаг.
Когда значения меняются местами, устанавливаем
флаг в значение True, чтобы повторить процесс
сортировки. Если перестановок не произошло,
флаг остаётся False и алгоритм останавливается.

````python
def bubble_sort(arr):
n = len(arr)
# Проходим по всем элементам массива
for i in range(n):
# Последние i элементов уже на месте
for j in range(0, n-i-1):
# Проходим по массиву от начала до `n-i-1`
# Меняем, если элемент найден больше, чем следующий
элемент
if arr[j] > arr[j+1]:
arr[j], arr[j+1] = arr[j+1], arr[j]
# Пример массива
arr = [64, 34, 25, 12, 22, 11, 90]
bubble_sort(arr)
print("Отсортированный массив:")
for i in range(len(arr)):
print("%d" % arr[i], end=" ")
````

# Бинарный поиск
 Бинарным (или двоичным) называют поиск
элемента упорядоченного множества через
многократное деление этого множества
пополам. Искомый элемент всегда будет
оказываться в одной из двух частей. Поиск
прекращается, когда обнаруживается
совпадение граничного элемента между
двумя разделенными блоками с заданным,
или когда заданный элемент не
обнаруживается вовсе.


````python
def binary_search(arr, x):
low,mid = 0
high = len(arr) - 1
while low <= high:
mid = (high + low) // 2
# Если x присутствует в середине
if arr[mid] < x:
low = mid + 1
# Если x меньше, игнорируем правую половину
elif arr[mid] > x:
high = mid - 1
# x находится в середине
else:
return mid
# Если мы достигли этой строки, элемента нет в массиве
return -1
````

# Срезы списков
Это механизм в Python, позволяющий извлекать подсписки из исходного списка с помощью специального синтаксиса. Срезы упрощают работу с последовательностями, делая код более компактным.

start- индекс начала среза(включительно). Если не указан, то беретсч с начала списка.
stop — индекс конца (не включается в результат).
Если не указан, то срез идёт до конца списка.
step — шаг, с которым берутся элементы. По
умолчанию равен 1.


# Срезы. Пропущенные границы
Если опустить start, Python берёт от начала списка. Если опустить stop - до конца. Если
опустить оба - берём весь список.
Для чего это используется:
1. Копирование списка: new_list = old_list[:]
Удобный способ сделать поверхностную копию.
2. Получение начала списка: lst[:n]
Например, взять первые 5 элементов.
3. Получение хвоста списка: lst[n:]
Например, убрать первые элементы и работать только с остатком.
4. Гибкость и краткость:вместо длинных циклов и условий можно быстро извлечь нужный
подсписок.


```
lst = [10,20, 30, 40, 50]
print(lst[:3]) #[10, 20, 30]
print(lst[2:]) #[30,40,50]
peint(lst[:]) # копия всего списка
```


# Срезы. Отрицателные индексы

Отрицательные индексы позволяют обращаться к элементам с конца списка. Это
удобно для работы с хвостами последовательностей.
```
lst = [1,2, 3, 4, 5]
print(lst[-2:]) #[4, 5]
print(lst[:-2]) #[1, 2 ,3 ]
peint(lst[-4:-1]) #  [2, 3 ,4 ]
```

# Срезы. Шаг в срезах
С помощью параметра step можно выбирать элементы через интервал или даже
переворачивать список.
```
lst = [0, 1, 2, 3, 4, 5, 6]
print(lst[::2]) #[0, 2, 4, 6]
print(lst[1:6:2]) #[1, 3, 5]
peint(lst[::-1]) # [6, 5, 4, 3, 2, 1, 0]
```
# Срезы и изменение списка

Срезы можно использовать не только для чтения, но и для изменения или удаления частей списка.

```
lst = [1, 2, 3, 4, 5]
lst[1:3] =[2-, 30, 40] # замена подсписка
print(lst) # [1,20,30,40,4,5]

lst[2:5] =[] # удаление подсписка
print(lst) # [1,20,5]
```

# Лямбда функции
Лямбда-функции в Python — это анонимные функции, которые определяются с помощью
ключевого слова lambda. В отличие от обычных функций, созданных через def, у лямбдафункции нет имени и она обычно записывается в одну строку. Синтаксис выглядит так:
lambda аргументы: выражение. Такая функция возвращает результат вычисления выражения
автоматически, без использования ключевого слова return.


Лямбда-функции чаще всего применяются в случаях, когда нужно описать простое
действие прямо «на месте», например, при сортировке (sorted), преобразовании
элементов (map), фильтрации (filter) или работе с данными в одну строку. Их основное
преимущество — компактность и удобство, когда нет необходимости создавать
полноценную именованную функцию. Однако для сложной логики лучше использовать
обычные функции с def, так как они более читаемы и понятны.



> Особенности лямбда-функций:
Они короче обычных функций и часто используются «в одну строку».
Удобны внутри функций высшего порядка (map, filter, sorted).
Обычно применяются там, где не нужно многократно вызывать
функцию.
Но если функция большая или сложная — лучше использовать def
для читаемости.
